```python
import pandas as pd
import plotly.offline as pyo
import plotly.figure_factory as ff
import numpy as np
```

```python
df = pd.read_csv('iris.csv')
df.head()
```

```python
df.info()
```

```python
df[df['class']=='Iris-setosa'].head(20)
```

```python
trace1 = df[df['class']=='Iris-setosa']['petal_length']
trace1.head()
```

```python
trace2 = df[df['class']=='Iris-versicolor']['petal_length']
```

```python
trace3 = df[df['class']=='Iris-virginica']['petal_length']
```

```python
hist_data = [trace1, trace2, trace3]
labels = ['Iris-setosa','Iris-versicolor','Iris-virginica']

fig = ff.create_distplot(hist_data,labels)#bin_size定义是optional
pyo.plot(fig, filename='Iris.html')
```
