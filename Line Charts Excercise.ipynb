```python
import pandas as pd
import numpy as np
import plotly.offline as pyo
import plotly.graph_objs as go
```

```python
df = pd.read_csv('2010YumaAZ.csv')
days=['TUESDAY','WEDNESDAY','THURSDAY','FRIDAY','SATURDAY','SUNDAY','MONDAY']
```

```python
df.head()
```

```python
df.info()
```

```python
data = []

for day in days:
    trace =go.Scatter(x=df['LST_TIME'] ,y=df[df['DAY']==day]['T_HR_AVG'],mode='lines',name=day)
    data.append(trace)
    
#data = [go.scatter(x=df['LST_TIME'],y=df[df['DAY']==day]['T_HR_AVG'],mode='lines',name=day) for day in df['DAY'].unique()] 
```

```python
data
```

```python
layout = go.Layout(title='Daily temp avgs')
fig = go.Figure(data=data,layout=layout)
pyo.plot(fig,filename='TempAvgs.html')
```
