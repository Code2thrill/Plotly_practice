```python
import plotly.offline as pyo
import plotly.graph_objs as go
import pandas as pd
```

```python
df = pd.read_csv("mpg.csv", na_values={'horsepower':'?'})
df.head()
```

```python
df.info()
```

```python
df.sort_values(by='horsepower' , axis=0,ascending=False,inplace = True)
df
```

```python
data = [go.Scatter(x=df['horsepower'],
                   y=df['mpg'],
                   text=df['name'],
                   mode='markers',
                   marker=dict(size=2*df['cylinders']))]#size都是试出来的
#notice to use go.Scatter() to creat a trace for a bubble chart
layout=go.Layout(title='Bubble Chart')
fig= go.Figure(data=data, layout=layout)
pyo.plot(fig,filename='Bubble Chart.html')
```

```python
data = [go.Scatter(x=df['horsepower'],
                   y=df['mpg'],
                   text=df['name'],
                   mode='markers',
                   marker=dict(size=df['weight']/100))]
#use bubble size to show the weight of these vehicles
layout=go.Layout(title='Bubble Chart')
fig= go.Figure(data=data, layout=layout)
pyo.plot(fig,filename='Bubble Chart2.html')
```

```python
data = [go.Scatter(x=df['horsepower'],
                   y=df['mpg'],
                   text=df['name'],
                   mode='markers',
                   marker=dict(size=df['weight']/100, color=df['cylinders']))]
#add color to the bubbles to show the cylinders of these vehicles, notice this is done in markers ()
layout=go.Layout(title='Bubble Chart')
fig= go.Figure(data=data, layout=layout)
pyo.plot(fig,filename='Bubble Chart3.html')
```

```python
data = [go.Scatter(x=df['horsepower'],
                   y=df['mpg'],
                   text=df['name'],
                   mode='markers',
                   marker=dict(size=df['weight']/100, color=df['cylinders'],showscale=True))]
#add scale of the colors; notice this is done in markers () 
layout=go.Layout(title='Bubble Chart')
fig= go.Figure(data=data, layout=layout)
pyo.plot(fig,filename='Bubble Chart4.html')
```
